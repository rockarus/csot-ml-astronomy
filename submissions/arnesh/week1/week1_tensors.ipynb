```python
import torch, torchvision, matplotlib, matplotlib.pyplot as plt
print ("The version of torch is", torch.__version__, ", the version of torchvision is", torchvision.__version__, "and the version of matplotlib is", matplotlib.__version__)
```

```python
print("Torch.cuda is available:", torch.cuda.is_available())
print("The device count is", torch.cuda.device_count())
if torch.cuda.is_available():
    print("The GPU is", torch.cuda.get_device_name())
```

```python
if torch.cuda.is_available():
  device = "cuda"
else:
  device = "cpu"
print("The device is", device)
```

```python
x=torch.randn(3,64,64)
print("Shape:", x.shape)
print("Datatype:", x.dtype)
print("Device:", x.device)
```

```python
x_gpu = x.to(device)
print("Device:", x_gpu.device)
print("Size:", x_gpu.shape)
```

```python
result = x_gpu[0] @ x_gpu[0]
print("Shape:", result.shape)
print("Device:", result.device)
```
